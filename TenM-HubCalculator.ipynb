````python
# TenM-Hub: Dynamic Pricing & Tenant Rating Model

import pandas as pd
import numpy as np

# --------------------------
# Dynamic Pricing Parameters
# --------------------------
PRICING_PARAMS = {
    'base_rates': {
        '1-5_units': 200,
        '6-10_units': 180,
        '11+_units': 170
    },
    'minimum_fee': 1500,
    'maximum_fee': 10000,
    'tenant_adjustments': {
        'excellent': 0.9,
        'good': 1.0,
        'average': 1.1,
        'poor': 1.3
    }
}

def get_base_rate(units):
    if units <= 5:
        return PRICING_PARAMS['base_rates']['1-5_units']
    elif units <= 10:
        return PRICING_PARAMS['base_rates']['6-10_units']
    else:
        return PRICING_PARAMS['base_rates']['11+_units']

def calculate_management_fee(units, tenant_quality):
    base_rate = get_base_rate(units)
    raw_fee = units * base_rate
    adjustment = PRICING_PARAMS['tenant_adjustments'].get(tenant_quality, 1.0)
    adjusted_fee = raw_fee * adjustment
    final_fee = max(min(adjusted_fee, PRICING_PARAMS['maximum_fee']), PRICING_PARAMS['minimum_fee'])
    return round(final_fee, 2)

# --------------------------
# Tenant Rating Model
# --------------------------
TENANT_WEIGHTS = {
    'payment_history': 0.4,
    'complaints': 0.2,
    'tenure_years': 0.2,
    'unit_condition': 0.2
}

def rate_tenant(payment_history, complaints, tenure_years, unit_condition):
    score = (
        payment_history * TENANT_WEIGHTS['payment_history'] +
        (1 - complaints) * TENANT_WEIGHTS['complaints'] +
        min(tenure_years / 5, 1.0) * TENANT_WEIGHTS['tenure_years'] +
        unit_condition * TENANT_WEIGHTS['unit_condition']
    )
    if score >= 0.8:
        return 'Good'
    elif score >= 0.5:
        return 'Medium'
    else:
        return 'High Risk'

# --------------------------
# Example Test Cases
# --------------------------

# Pricing calculation test
print("Fee (8 units, good tenants):", calculate_management_fee(8, 'good'))

# Tenant scoring test
print("Tenant risk profile:", rate_tenant(payment_history=0.9, complaints=0.1, tenure_years=3, unit_condition=0.8))
````
